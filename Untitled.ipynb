```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv('C://MachineLearningData//ufc-master.csv',header=None)
```

```python
df.head()
```

```python
# remplacer les noms de colonnes par la premiere ligne
new_header = df.iloc[0] 
df = df[0:] 
df.columns = new_header  
```

```python
# supprimer la premiere ligne
df.drop(0,0,inplace=True)
```

A la recherche de données manquantes

```python
# Identifions les colonnes qui ont des données manquantes
cols = df.columns[:60] # les 60 premieres colonnes
colours = ['#000099', '#ffff00'] # jaunes: valeurs manquantes -  bleu non manquantes
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# recherche du poucentage de données manquantes par colonnes
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# presentation d'un histogramme 
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing


# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in df.columns if 'ismissing' in col]
df['num_missing'] = df[ismissing_cols].sum(axis=1)

df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

```python
# impute the missing values and create the missing value indicator variables for each numeric column.
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values

for col in numeric_cols:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  # only do the imputation for the columns that have missing values.
        print('imputing missing values for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing
        med = df[col].median()
        df[col] = df[col].fillna(med) 
```
